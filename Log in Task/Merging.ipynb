## Merging

```
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import copy
import datetime 
from collections import Counter
```

```
boxOffice = pd.read_csv("boxoffice_weekly_gross_movie_filled_seasonality.csv")
boxOffice['startdate'] =  pd.to_datetime(boxOffice['startdate'])
boxOffice['enddate'] =  pd.to_datetime(boxOffice['enddate'])
print(boxOffice.shape)
boxOffice.head(5)
```

```
filled_set = pd.read_csv("Film-masterfile (from 'film-merge-8th.dta')-filled.csv")
# turn release time into timestamp
filled_set['releasedate'] =  pd.to_datetime(filled_set['releasedate'])
filled_set['closedate'] =  pd.to_datetime(filled_set['closedate'])
filled_set['year_week_average_gross'] = np.nan
filled_set['year_week_average_totalmovies'] = np.nan
filled_set.head(5)
```

```
for i in range(filled_set.shape[0]):
  for j in range(boxOffice.shape[0]):
    if filled_set['releasedate'][i] >= boxOffice['startdate'][j]:
      # fixed week
      filled_set['year_week_average_gross'][i] = boxOffice['year_week_average_gross'][j]
      filled_set['year_week_average_totalmovies'][i] = boxOffice['year_week_average_totalmovies'][j]
      break
  # process
  if i%200 == 0:
    print(i)
filled_set.head(10)
```

```
filled_set.to_csv("filled_set_1008.csv")
```
